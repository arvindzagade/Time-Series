# Customised Plots

```python
df2 = pd.read_csv('df2.csv')
```

```python
df2.plot.line()
```

```python
df2.plot()
```

```python
df2['c'].plot.line(figsize=(20,5),ls='-.',c='black',lw=5)
```

### Adding Title and lables on graph

```python
title = 'My plot title'
xlabel = 'My x Data'
ylabel = 'My y Data'
```

```python
ax = df2['c'].plot.line(figsize=(10,5),ls='-.',c='black',lw=5,title =title)
ax.set(xlabel=xlabel,ylabel=ylabel)
```

## adding legends

```python
ax = df2.plot()
ax.legend(loc=0)
```

```python
## Moving label to outside of the box
ax = df2.plot()
ax.legend(loc=0,bbox_to_anchor=(1.0,1.0))
```
